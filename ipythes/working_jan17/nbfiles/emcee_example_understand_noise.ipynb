```python
import numpy as np

# Choose the "true" parameters.
m_true = -0.9594
b_true = 4.294
f_true = 0.534

# Generate some synthetic data from the model.
N = 50
x = np.sort(10*np.random.rand(N))
yerr = 0.1+0.5*np.random.rand(N)
```

```python
plt.plot(x,np.random.rand(N),'.',color="blue") #uniform distribution [0,1)
plt.plot(x,0.5*np.random.rand(N),'.',color="red") #scaled uniform distribution [0,.5)
plt.plot(x,0.1+0.5*np.random.rand(N),'.',color="green") #scaled and shifted uniform distribution [0.1,0.6)
```

```python
plt.plot(x,np.random.randn(N),'o') #sample from gaussian dist. mean = 0 var = 1
```

```python
plt.plot(x,y,'.',color='red') # true model
plt.plot(x,np.abs(f_true*y),'.',color='blue') #abs value of fractional underestimate of true model
plt.plot(x,np.abs(f_true*y) * np.random.randn(N), '.', color='green') #abs value of frac. underest. mult by gaussian
plt.plot(x, (y + np.abs(f_true*y) * np.random.randn(N)), 'o',color='purple')
```
